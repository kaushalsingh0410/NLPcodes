```python
curent_path = pathlib.Path(r'E:\kaushal\data science\nlp\code\campusx\TRANING\AGameOfThrones')
curent_path
```

```python
lemma = WordNetLemmatizer()
```

```python
corpus = []
for file in curent_path.glob('*'):
    for line in file.read_text().split('\n'):
        # print(simple_preprocess(line))
        corpus.append([lemma.lemmatize(word) for word in simple_preprocess(line) if word not in stopwords.words('english')])
```

```python
corpus
```

```python
wtv = Word2Vec(window = 5,min_count = 2)
```

```python
wtv.build_vocab(corpus)
```

```python
dir(wtv)
```

```python
wtv.train(corpus,total_examples = wtv.corpus_count,epochs = wtv.epochs)
```
